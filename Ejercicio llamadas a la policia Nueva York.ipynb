```python
#Si queremos visualizar la localizaci√≥n de los sucesos, utilizamos las coordenadas y un scatter:
req_df.plot(kind='scatter', x='Longitude', y='Latitude', alpha=.1, figsize=(12, 12), s=1)
```

```python
req_df[req_df['Borough']=='MANHATTAN'].plot(kind='scatter', x='Longitude', y='Latitude', alpha=.1, figsize=(12, 12), s=1)
```

```python
req_df['Borough'].value_counts()
colors = ('lime', 'orange', 'm', 'b', 'pink', 'peru')
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
for color, barrio in zip(colors, req_df['Borough'].value_counts().index):
    req_df[req_df['Borough']==barrio].plot(kind='scatter', c=color, label=barrio, ax=ax, x='Longitude', y='Latitude', alpha=.1, s=1)
plt.legend()
```

```python
#D√≥nde se concentran m√°s las llamadas? Para ver eso, vamos a teselar el plano, haciendo una rejilla:
req_df.plot(kind='hexbin',x='Longitude', y='Latitude', gridsize=80, figsize=(12, 8), colormap='jet')
```

```python
#Para descartar zonas enlas que no hay nada de quejas:
req_df.plot(kind='hexbin',x='Longitude', y='Latitude', gridsize=80, figsize=(12, 8), colormap='jet', mincnt=1)
```
