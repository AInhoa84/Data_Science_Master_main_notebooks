```python
#Ejercicio: implementar algo que nos diga, mediante una pila, si una palabra es palíndroma o no (palín-
#droma significa que se lee igual del derecho que del revés).Hay que implementar una función que reciba un
#string y nos devuelva un booleano:
#MI SOLUCIÓN (QUE NO FUNCIONA):
def palindroma(a:str, *args) -> bool:
    before = []
    after = []
    for i in a:
        before.append(i)
        after.append(a.pop())
    if before == after [::-1]:
        return True
    else:
        False

palindroma('ada')
```

```python
#Solución del profesor:
def is_palindrome(s:str) -> bool:
    stack = list(s)
    reverse = ''
    while len(stack):
        reverse += stack.pop()
    return s == reverse
is_palindrome('alan')
```

X = np.arange(1, 10, .05).reshape(-1, 1)
Y = np.array([x + np.random.normal(0, x/5) for x in X]).reshape(-1, 1)

Visualizar los puntos con un scatter plot.
Calcular la pendiente y el sesgo (bias o intercept) mediante la fórmula analítica de la regresión lineal sobre el dataset.

Pendiente (Beta 1): https://wikimedia.org/api/rest_v1/media/math/render/svg/f3d693b14cf10eab3c9930d3c7020b2697befb0b

Sesgo (Beta 0):
https://wikimedia.org/api/rest_v1/media/math/render/svg/2291a51c3be5577498e10d1b610f435ec03da0ce

Predecir y (en una nueva columna y_pred) en base a los valores calculados siguiendo la fórmula de la regresión lineal: y = Beta1 * X + Beta0.

Visualizar la recta de regresión junto a los puntos anteriores.

```python
import numpy as np
import pandas as pd
#Generación del dataframe:
X = np.arange(1, 10, .05).reshape(-1, 1)
Y = np.array([x + np.random.normal(0, x/5) for x in X]).reshape(-1, 1)
dataframe = {'X': X.squeeze(), 'Y': Y.squeeze()}
df = pd.DataFrame(dataframe)
df
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
#Visualización de los puntos mediante scatter plot:
df.plot(kind='scatter', x='X', y='Y')
```

```python
#Calculamos la pendiente usando la fórmula que aparece en el link y programándola:
numerador = ((df['X'] - df['X'].mean()) * (df['Y'] - df['Y'].mean())).sum()
denominador = ((df['X'] - df['X'].mean()) **2).sum()
pendiente = numerador/denominador
print('La pendiente es: {:.2f}'.format(pendiente))
```

```python
#Calculamos el sesgo o el bias:
bias = df['Y'].mean() - pendiente * df['X'].mean()
print('El sesgo es: {:.2f}'.format(bias))
```

```python
#Ploteamos los puntos:
ax = df.plot(kind='scatter', x='X', y='Y')
ax.plot([df['X'].min(), df['X'].max()], [pendiente * df['X'].min() + bias, pendiente * df['X'].max() + bias],
       lw=4, color='r')
```

```python
#Cálculo de la media móvil de un dataset:
import numpy as np
from matplotlib import pyplot as plt
#Nos ceramos un dataset
X = np.arange(0, 1, .005)
y = 2 * X + np.random.normal(.2, X)
fig, ax = plt.subplots(1, 1)
ax.scatter(X, y)
plt.tight.layout()
```

```python
import pandas as pd
df = pd.DataFrame({'y':y})
#Calculamos la media movil:
df['moving_avg'] = df.rolling(window=10).mean()
fig, ax = plt.subplots(1, 1)
ax.scatter(X, y)
ax.plot(X, df['moving_avg'], c='r', lw=4)
plt.tight.layout()
```

```python
#Podemos ampliar la ventana o el periodo para poder ver más claramente y quitar el ruido:
for i in [2, 5, 10, 20, 50, 100]:
    df['moving_avg_{}'.format(i)] = df['y'].rolling(window=i).mean()
fig, ax = plt.subplots(1, 1)
ax.scatter(X, y)
for col in df.columns:
    if col.startswith('moving_avg_'):
        ax.plot(X, df[col], lw=2, label=col)
plt.legend()
plt.tight_layout()
```

```python
#Vamos a modificar un poco el dataset para verlo de otra manera:
import numpy as np
X = np.arange(-1, 1, .005)
y = 2 * X **2 + np.random.normal(.2, abs(X))
df = pd.DataFrame({'y':y})
#Podemos ampliar la ventana o el periodo para poder ver más claramente y quitar el ruido:
for i in [2, 5, 10, 20, 50, 100]:
    df['moving_avg_{}'.format(i)] = df['y'].rolling(window=i).mean()
fig, ax = plt.subplots(1, 1, figsize =(15, 7))
ax.scatter(X, y)
for col in df.columns:
    if col.startswith('moving_avg_'):
        ax.plot(X, df[col], lw=2, label=col)
plt.legend()
plt.tight_layout()
```
